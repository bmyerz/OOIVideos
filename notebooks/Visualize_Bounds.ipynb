```python
import os
```

```python
video_path = os.path.join(os.getcwd(),'..','results','bounds_rolling_variance')
```

```python
import glob
```

```python
listOfFiles = os.listdir(video_path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
fig = plt.figure(figsize = (20,5))
height = 1
fig.canvas.draw()
for item in listOfFiles:
    if '0101' in item:
        bounds = pd.read_csv(os.path.join(video_path,item))
        plt.plot(bounds['LB'],height*np.ones(len(bounds['LB'])),'ro')
        plt.plot(bounds['UB'],height*np.ones(len(bounds['LB'])),'bo')
        height = height + 1
        plt.ylim([0,10])
plt.title('20160101 Scene Bounds'  )

```
